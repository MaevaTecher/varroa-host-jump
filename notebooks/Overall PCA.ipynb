# Exploratory analyses

```r
library(ggplot2)
library(ggrepel)
library(dplyr)
```

```r
mat <- read.table("../data/var/filtered.mat", skip=1, header=F, row.names=1)
mat$id <- colnames(mat) <- rownames(mat)
dim(mat)
mds <- cmdscale(as.dist(mat), eig = T)
```

```r
plot(mds$eig)
```

```r
pca <- as.data.frame(mds$points)
colnames(pca) <- c("PCA1", "PCA2")
pca$id <- rownames(mds$points)
pca <- left_join(pca,metadata)
ggplot(pca, aes(PCA1, PCA2, color = host, shape = species, label = id))+geom_point()+geom_text_repel() 
```

```r
ggplot(pca, aes(PCA1, PCA2, color = host, shape = species, label = id))+geom_point()+geom_text_repel()+xlim(.1,.2)+ylim(-.025,.025)
```
